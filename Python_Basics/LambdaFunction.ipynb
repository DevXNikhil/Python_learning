```python
# Normal Function which calculates power
def test(a,b):
    return a**b
```

```python
test(2,2)
```

```python
# Lambda Function (Anonymous Function)
# Example - 1 Calculating Power
power = lambda a,b : a**b
```

```python
power(3,2) # Function Calling
```

```python
# Example - 2 Addition of Two numbers
add = lambda a,b: a+b
```

```python
add(3,5)
```

```python
# Example - 3 Multiplication of Two numbers
multi = lambda c,d : c*d
```

```python
multi(5,5)
```

```python
# Example - 4 Modulus
mod = lambda j,k : j%k
```

```python
mod(4,40)
```

```python
# Example - 5 Convert Celsius to Fahrenheit (°F=(25×9/5)+32=45+32=77°F)
c_to_f = lambda c : (9/5)*c + 32
```

```python
c_to_f(25)
```

```python
# Example - 6 Find maximum amoung two numbers
maximum = lambda a,b : a if a>b else b
```

```python
maximum(10,100)
```

```python
# Example - 7 Find length of string
s = "pwskills"
l_o_s = lambda a : len(a)
```

```python
l_o_s(s)
```
