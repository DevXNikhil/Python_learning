```python
# Example - 1 For Class Method
# Overloading of a constructor with the help of class_method
class pwskills1:
    def __init__(self,name,mail): # Constructor for value initialization
        self.name = name
        self.mail = mail

    # Overloading -> # IMP -> We are Overloading -> __init__ with the help of class method
    @classmethod  # -> It is an inbuilt decorator in Python
    def student_info(cls,name,mail): # Class Method -> CLS is a pointer -> Cls refers to the class
        return cls(name,mail)

    def st_details(self): # Display Function
        print("Student Name is : "+self.name)
        print("Student E-mail is : "+self.mail)
```

```python
pws = pwskills1.student_info("Nikhil","DevX@gmail.com") # Not Creating an Object, directly Using class's Name 
```

```python
pws.name
```

```python
pws.mail
```

```python
pws.st_details()
```

```python
# Example - 2 For Class Method
# Overloading of a constructor with the help of class_method
# Changing mobile number and accessing variable of a class -> Mobile_num
class pw:
    mobile_num = 77895479   # Class Variable
    
    def __init__(self,name): # Constructor
        self.name = name

    @classmethod             # class method
    def change_mobile_num(cls,mobile_num): # Function for changing mobile number 
        pw.mobile_num = mobile_num
        return pw.mobile_num
        
    @classmethod             # class method
    def student(cls,name):   # IMP -> We are Overloading -> __init__ with the help of class method
        return cls(name)
        
    def display(self):          # Display Function, here we have to pass (self) -> class reference 
        print("Name of student is : ",self.name)
        print("Mobile Number is : ",int(pw.mobile_num))
```

```python
pw = pw.student("ZORO")
```

```python
pw.name
```

```python
pw.mobile_num # Student Contact is also accessible without making any object
```

```python
pw.change_mobile_num(6786543) # Changing mobile_number
```

```python
pw.display()
```

```python
# Example - 3 For Class Method
# How can we include an external function as a method in a class.
class pw1:
    
    def __init__(self,course): # Constructor
        self.course = course

    @classmethod               # Overloading using class_method
    def student_course(cls,course):
        return cls(course)

    @classmethod               
    def display_course(self):
        print("Course : ",self.course)
```

```python
def student_detail(cls,student_name): # 1st External Function 
    print("Student name is : ",student_name)
```

```python
pw1.student_detail = classmethod(student_detail) # Including 1st external function to class -> pw1
```

```python
pw1.student_detail("Nikhil") # Accessing external function using pw1
```

```python
pw1.student_detail("Kakashi")
```

```python
def mentor(cls,list_of_mentor): # 2nd External function
    print(list_of_mentor)
```

```python
pw1.mentor = classmethod(mentor) # Including 2nd external function to class -> pw1
```

```python
pw1.mentor(["Sudhanshu sir","krish sir"]) # Function calling
```

```python
# Example - 4 Deleting or deactivating of a class's method,function
class pw20:
    mobile_num = 77895479   # Class Variable
    
    def __init__(self,name): # Constructor
        self.name = name

    @classmethod             # class method
    def change_mobile_num(cls,mobile_num): # Function for changing mobile number 
        pw.mobile_num = mobile_num
        return pw.mobile_num
        
    @classmethod             # class method
    def student(cls,name):   # IMP -> We are Overloading -> __init__ with the help of class method
        return cls(name)
        
    def display(self):          # Display Function, here we have to pass (self) -> class reference 
        print("Name of student is : ",self.name)
        print("Mobile Number is : ",int(pw.mobile_num))
```

```python
pw20.change_mobile_num(678639)
```

```python
del pw20.change_mobile_num # Deleting function -> change_mobile_num
```

```python
pw20.change_mobile_num(5455334) # Function is deleted
```

```python
# 2nd Method of Deleting or deactivating of a class's method,function
# Using function -> delattr() -> delete attribute
delattr(pw20,"student") # Function student is deleted 
```

```python
pw20.student() # Function student is deleted 
```

```python
# Deleting variable -> mobile_number
delattr(pw20,"mobile_num")
```
