# Encapsulation - Wrapping data and methods into a single unit and restricting direct access to data.

```python
class test:
    def __init__(self , a , b): # Constructor
        self.a = a
        self.b = b
```

```python
t = test(3,2)   # making object and initializing values 
```

```python
# Example - 1 of Encapsulation
class car:
    def __init__(self,model,speed,brand):
        self.__model = model  # __ this underscore used to hide this variable by accessing it externally
        self.__speed = speed
        self.__brand = brand
```

```python
c = car("M4",300,"BMW")
```

```python
c.model # here we are not getting output because variables are (Hidden) using __ 
```

```python
c.speed 
```

```python
c._car__model # But using this thing we can access the m
```

```python
c._car__speed
```

```python
c._car__brand
```

```python
# Example - 2
class car1 :
    def __init__(self,year,model,speed):
        self.__year = year # hidden
        self.__model = model # hidden
        self.__spedd = speed # hidden
  
    def set_speed(self,speed): # Instead of accessing variable directly we can call this function
        self.__speed = 0 if speed < 0 else speed

    def get_speed(self): # And for getting that value we can call this function
        return self.__speed
```

```python
c = car1(2022,"M5",250)
```

```python
c.set_speed(200)
```

```python
c.get_speed()
```

```python
# Example - 4 
class bank_account:
    def __init__(self,balance):
        self.__balance = balance # THIS BALANCE IS HIDDEN 
        
    def deposit(self,amount): # Money Deposit Function
        self.__balance = self.__balance + amount

    def withdraw(self,amount): # Money Withdraw Function
        if self.__balance >= amount : # Condition
            self.__balance = self.__balance - amount
            return True
        else :
            return False
          
    def get_balance(self): # Function for showing remaining amount
            return self.__balance
```

```python
b2 = bank_account(500000) # We are passing this value for initialization
```

```python
b2.deposit(500000) # Deposit
```

```python
b2.withdraw(5000) # Withdrawing
```

```python
b2.get_balance() # Getting Remaining Balance wihtout accessing variable - balance
```

```python
b2.withdraw(1500000)
```
