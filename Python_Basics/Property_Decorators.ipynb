```python
# Property Decorators
```

```python
# Example - 1
class pwskills:
    def __init__(self,course_price,course_name):
        self.__course_price = course_price # This variable is protected,private
        self.course_name = course_name

    @property
    def course_price_access(self):
        return self.__course_price

    # setter
    @course_price_access.setter # setting the price of course 
    def course_price_set(self,price):
        if price <= 3500:
            pass
        else:
            self.__course_price = price

    
    @course_price_access.deleter
    def delete_course_price(self):
        del self.__course_price
```

```python
pw = pwskills(3500,"Data science")
```

```python
pw.__course_price # course_price variable is not accessible 
```

```python
pw._pwskills__course_price # Accessing using class name
```

```python
pw.course_price_access # Now it is accessible using method and property decorator
```

```python
pw.course_price_set = 4500 # setting price 
```

```python
pw.course_price_access  # Accessing price
```

```python
del pw.delete_course_price # Deleting course_price attribute
```

```python
pw.course_price_access # not accessible
```
