```python
# Example - 1
class pwskills:
    def __init__(self,course_price,course_name):
        self.__course_price = course_price # This variable is protected,private
        self.course_name = course_name

    # Property Decorator
    @property
    def course_price_access(self):
        return self.__course_price

    # Setter
    @course_price_access.setter # setting the price of course 
    def course_price_set(self,price):
        if price <= 3500:
            pass
        else:
            self.__course_price = price

    # Deleter
    @course_price_access.deleter
    def delete_course_price(self):
        del self.__course_price
```

```python
pw = pwskills(3500,"Data science")
```

```python
pw.__course_price # course_price variable is not accessible becuase that is hidden or protected
```

```python
pw._pwskills__course_price # Accessing using class name
```

```python
pw.course_price_access # Now it is accessible using method and property decorator
```

```python
pw.course_price_set = 4500 # Setting the price 
```

```python
pw.course_price_access  # Accessing the price
```

```python
del pw.delete_course_price # Deleting course_price attribute
```

```python
pw.course_price_access # it will give -> error , not accessible
```

```python
# Example - 2
# Property Decorators
# Setter
# deleter
class bank_account:
    def __init__(self,name,balance,account_num): 
        self.name = name
        self.__balance = balance  # Private attribute
        self.__account_num = account_num # Private attribute
        
    @property
    def bank_balance_access(self):
        return self.__balance

    @bank_balance_access.setter # Setter function
    def balance_deposit(self,balance):
        if balance <=0 :
            print("You cannot deposit an amount of 0 or less in the bank.")
        else :
            self.__balance = self.__balance + balance
    
    @bank_balance_access.deleter # Deleter function
    def delete_balance(self):
        del self.__balance
```

```python
b1 = bank_account("XYZ",5000,9340000084373) # Object creation
```

```python
b1.name
```

```python
b1.balance # We can not access private varibale 
```

```python
b1.account_num
```

```python
b1._bank_account__balance # Single underscore for _class
                            # Double underscore for __private variable
```

```python
b1._bank_account__account_num
```

```python
b1.bank_balance_access # Accessing private variable using property decorator
```

```python
# 1 Setter function
b1.balance_deposit = 5000 # depositing money into bank = 5000
```

```python
b1.bank_balance_access # Intial bank balance = 5000
                        # deposit money = 5000
```

```python
b1.balance_deposit = 0
```
