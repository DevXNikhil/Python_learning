```python
# Example - 1 for static method
class pwskills1:
    def student(self,name,course):
        return self.name , self.course
    @staticmethod
    def course(course_list):
        print(course_list) 
```

```python
# Without creating an obj we can access the function of a class using static method
pwskills1.course(["Data Science ","Java spring boot"]) # No obj is created
```

```python
# Example - 2 for using @staticmethod inside the class with the help of @classmethod
class pwskills2:
    def student(self,name,mail): # Function
        return self.name,self.mail
        
    @staticmethod # Static method
    def mentor_class(list_mentor):
        print(list_mentor)

    @classmethod # Class method
    def class_name(cls): # Using a @staticmethod within the same class with the help of [@classmethod]
        cls.mentor_class(["Sudhashu sir","Krish sir"]) # cls refers to the class_name, cls = pwskills2
```

```python
pwskills2.class_name() # Accessing class_name without object
```

```python
 # Example - 2 for using @staticmethod inside the class with the help of @classmethod
# So for every object we made @Static method only take memory at once not every time that's why we use static method
class pwskills2:
    def student(self,name,mail): # Function
        return self.name,self.mail

    @staticmethod # static method
    def sub_of_mentor(subject):
        print(subject)

    @staticmethod # Static method
    def mentor_mail_id(mail_id):
        print(mail_id)
        
    @staticmethod # Static method
    def mentor_class(list_mentor):
        print(list_mentor)

    @classmethod # Class method
    def class_name(cls): # Using a @staticmethod within the same class with the help of [@classmethod]
        cls.mentor_class(["Sudhashu sir","Krish sir"]) # cls refers to the class_name, cls = pwskills2
        cls.mentor_mail_id("Sudhanshu@gmail.com")  # Accessing mentor_mail_id method
        cls.sub_of_mentor("Python") # Accessing sub_of_mentor method
```

```python
pwskills2.class_name()
```

```python
# Example - 3 # Calling static method using class method,or static method
# SM - Static method
class pwskills3:
    def student(self,name,mail): # Function
        return self.name,self.mail

    @staticmethod # Static method # this function is called by the mentor_class function
    def mentor_mail_id(mail_id):
        print(mail_id)
        
    @staticmethod # Static method
    def mentor_class(list_mentor):
        print(list_mentor)
        pwskills3.mentor_mail_id("Sudhanshusir@gmail.com") # calling mentor_mail_id(SM) inside mentor_class (SM)

    @classmethod # Class method
    def class_name(cls): # Using a @staticmethod within the same class with the help of [@classmethod]
        cls.mentor_class(["Sudhashu sir","Krish sir"]) # cls refers to the class_name, cls = pwskills2

    # This function will need an object to be called 
    def mentor(self,mentor_list):
        print(mentor_list)
        self.mentor_class(["Avinash sir","Dhurv sir"]) 
```

```python
# 1st Method of calling 
pwskills3.class_name() # Calling using class_method
```

```python
# 2nd Method of calling 
pwskills3.mentor_class(["Sudhanshu sir ","Krish sir"]) # Directly calling using static method -> mentor_class
```

```python
pw = pwskills3() # Object Creation as pw
```

```python
# 3rd method of calling 
# Object - pw
pw.mentor(["Avinash sir","Dhurv sir"]) # Calling through an object -> to a @staticmethod mentor_class
```
