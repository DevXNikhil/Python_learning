# question 1 - sum up the number


```python
n = int(input("Enter a number : "))
starting_point = 0
counter = 1

while counter <= n:
    starting_point = starting_point + counter
    counter += 1
print(starting_point)
```

# question 2 - Find factorial of a number

```python
number = int(input("Enter a number :"))
factorial = 1

while number>0:
        factorial = factorial * number
        number = number - 1
print(factorial)
```

# question 3 - Fibonacci series - Addition of two previous number
# 0 1 1 2 3 5 8 13 

```python
num = int(input("Enter a number :"))
a , b  = 0 ,1 
counter = 0
 while counter < num :
    print(a)
    c = a+b
    a = b
    b = c
    counter = counter + 1      
```

# question 4 - Reverse a string 

```python
word = input("Enter a string which you want to reverse : ")
rev = ""

le = len(word) - 1

while le >= 0 :
    rev = rev + word[le] 
    le = le - 1
    
print("Reversed string is : ",rev)
```

# Simple Method of reversing a string

```python
s = "hello"
print(s[::-1]) # reversing
print(s[::1])  # printing 
```
