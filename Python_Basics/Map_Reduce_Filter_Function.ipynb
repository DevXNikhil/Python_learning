```python
# Example - 1
l = [3,4,5,6,6]
def sq(x):    # Function
    return x**2
```

```python
list(map(sq, l)) # Using Map with Normal Function
```

```python
# Example - 2
l = [5,6,7,8,9]
list(map(lambda x : x**2,l)) # Using Map with Lambda Function
```

```python
# Example - 3
l1 = [1,3,35,46,5]
l2 = [1,2,3,4,5]
def add(a,b): # Using Map with Normal Function
    return a + b
```

```python
list(map(add,l1,l2))
```

```python
# Example - 4
l1 = [1,3,35,46,5]
l2 = [1,2,3,4,5]
list(map(lambda x,y : x + y ,l1,l2)) # Using Map with Lambda Function 
```

```python
# Example - 5
s = "pwskills"
list(map(lambda s:s.upper(),s))
```

```python
# Example - 6
s = "PWSKILLS"
list(map(lambda s : s.lower(),s))
```

```python
from functools import reduce # Reduce is available in function tool package,we need to import.
```

```python
# Example - 1
l = [1,2,3,3,4,5]
reduce(lambda x,y:x+y,l)
```

```python
# Example - 2
l = [1,2,3,4,5,6]
print(reduce(lambda x,y:x*y,l)) # Here number of arguments always should be 2 not 3
```

```python
reduce(lambda x,y : x+y,[]) # It is not gonna work
```

```python
reduce(lambda x,y : x+y,[1]) # Exception case
```

```python
# Example - 3
l = [1,2,3,4,5,6,50,]
reduce(lambda x,y : x if x>y else y ,l)
```

```python
# Example - 1 Find Even numbers
l = [243,4,5,6,434,233,32,123,45,66,78,90,100]
list(filter(lambda x : x%2==0,l))
```

```python
# Example - 2 Find numbers which are divisible by 5
l = {5,25,30,35,36,38,90,22,11,12,25} # This is a set so we will only get unique values
list(filter(lambda x : x%5 == 0,l))
```

```python
# Example - 3 Find Odd numbers
l = [243,4,5,6,434,233,32,123,45,66,78,90,100]
list(filter(lambda x : x%2!=0,l))
```

```python
# Example - 4 Find negative numbers
l = [-1,-4,-6,6,7,8,4,2,-10,]
list(filter(lambda x : x<0 ,l))
```

```python
# Example - 5 Find a string which has length 5 only
s = ["pwskills","hello","bruhh","Python","Data_science"]
list(filter(lambda s : len(s)<6,s))
```
