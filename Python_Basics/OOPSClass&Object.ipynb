```python
# Example - 2 
class Pwskills1 :

    def __init__(self,contact,mail,student_id,): # Constructor made using __init__ keyword
        self.contact = contact
        self.mail = mail
        self.student_id = student_id

    def return_student_details(self):  # Method(Function) for returning details 
        return self.contact ,self.mail,self.student_id
```

```python
rohan = Pwskills1() # It will give error if nothing passed in arguments as details of student
```

```python
rohan = Pwskills1(1001000082,"abc@gmail.com",101) # 1st Object of Pwskills(class) for rohan
```

```python
rohan.return_student_details() # calling method for showing details of rohan
```

```python
rohan.contact
```

```python
# A class can have multiple objects
```

```python
gaurav = Pwskills1(647534687,"gaurav@gmail.com",102) # 2nd Object of Pwskills(class) for gaurav
```

```python
gaurav.return_student_details() # calling method for showing details of gaurav
```

```python
gaurav.student_id
```

```python
gaurav.contact
```

```python
gaurav.mail
```

```python
# Example - 4
class car :

    def __init__(car,Engine_cc,color,price):
        car.Engine_cc = Engine_cc
        car.color = color
        car.price = price
    def display(car):
        return car.Engine_cc,car.color,car.price
```

```python
bmw = car(2993,"BLACK",15000000) # 1st Object
```

```python
bmw.color
```

```python
bmw.Engine_cc
```

```python
bmw.price
```

```python
bmw.display()
```

```python
mustang = car(4951,"Yellow",10000000) # 2nd Object
```

```python
mustang.price
```

```python
mustang.color
```

```python
mustang.Engine_cc
```

```python
mustang.display()
```
