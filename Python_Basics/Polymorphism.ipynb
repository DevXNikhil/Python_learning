# Polymorphism - means one thing can work in different ways.

```python
# Example - 1 
def test(a,b): 
    return a + b
```

```python
test(4,5) # Here we are calling test function for Addition
```

```python
test("Nikhil ","Gour") # Here we are calling test function for concatination
```

```python
test([3,4,5],[6,7,8])  # Here we are calling test function for list concatination
```

```python
# Example - 2
class data_science: # Class 1 

    def syllabus(self): # Method
        print("Syllabus of data science")
```

```python
class web_dev:  # class 2

    def syllabus(self): # Method
        print("Syllabus of web development")
```

```python
def class_parcer(obj): # External Function for calling
    for i in obj:
        i.syllabus()
```

```python
data_science = data_science() # Object Creation
```

```python
web_dev = web_dev() # Object Creation
```

```python
obj = [data_science,web_dev] # Assigning list into obj
```

```python
class_parcer(obj) # calling class_parcer for both objects
```

```python
# Example - 3
class vehicle1:
    def veh(self):
        print("This is a Motorbike")
```

```python
class vehicle2:
    def veh(self):
        print("This is a car")
```

```python
def display(objec):
    for i in objec:
        i.veh()
```

```python
bike = vehicle1()
```

```python
car = vehicle2()
```

```python
objec = [bike,car]
```

```python
display(objec)
```
