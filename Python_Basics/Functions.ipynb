```python
# Question - 1 Write a function to extract only int and float values from a list.
def test7(l):
    l1 = []
    for i in l:
         if type(i) == int or type(i) == float:
            l1.append(i)
    return l1
```

```python
# Question - 2 Write a function to flatten a list and extract all int and float values.
def test8(a):
    """ Function to extract num data from list""" # docstring 
    l1 = []
    l2 = []
    for i in a:
        if type(i) == list:
            for j in i:
                l1.append(j)  
        else:
            if type(i) == int or type(i) == float:
                l1.append(i)
            l2.append(i)  
    return l1
```
