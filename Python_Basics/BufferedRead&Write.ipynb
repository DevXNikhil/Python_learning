```python
# Buffered reading and writing 
# Example - 1 
# Writing data into file
import io
```

```python
with open("test.txt","wb")as f: # wb = write binary
    file = io.BufferedWriter(f)

    file.write(b"Hello\n") # Writing data into file
    file.write(b"Bruh")   
    file.flush() # Closing the file
```

```python
# Example - 2
# Reading a file
with open("test.txt","rb") as f:
    file = io.BufferedReader(f)
    print(file.read())
    print(file.read(2)) # We can define byte here () 3,4,5,6
```

```python
# Example - 3
with open("test.txt","rb") as f:
    file = io.BufferedReader(f)
    print(file.read(2)) # Read 2 byte only
```

```python
# Example - 4 
# Writing data 
with open("test1.txt","wb") as f:
    file = io.BufferedWriter(f)

    file.write(b"Just do it\n")
    file.write(b"dont worry about future")
    file.flush()
```
